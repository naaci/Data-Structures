This is a simple client application using Python `socket` module in standart library. We start by importing that.

```python
import socket
```

To use the socket module first we are creating a socket object by calling:

```python
server = socket.socket()
```

```python
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
server.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEPORT, 1)
```

## Creating A Server Application

Secondly, we need the address of the **server** where the address is a tuple of two items:
- 1st is a string that is the ip address or domain name to be served on.
- 2nd is an integer **chosen** between $0$ ond $2^{16}-1$ that will be the port number.

```python
address = '172.16.2.5', 8080
address = '', 9001
```

If you want only to allow only yourself (another application running on the same computer) to connect then you can use `'127.0.0.1'` as IP address or `'localhost'` for domain name.

```python
server.bind(address)
```

To start listening that address:

```python
server.listen()
```

To start accepting the connections call:

```python
try:
  s, _ = server.accept()
except KeyboardInterrupt: pass
```

Now `s` is another socket that can communicate with the remote.

The remote is supposed to sent a `'Hello, world!'` mesage. Let's check it.

```python
if s.recv(2048).decode() != 'Hello, world!':
  s.shutdown(socket.SHUT_WR)
else:
  print("Connection is successful")
  while True:
    try:
      s.sendall(input().encode())
      print(s.revc().decode())
    except KeyboardInterrupt: break
    finally:
      s.shutdown(socket.SHUT_WR)
```
