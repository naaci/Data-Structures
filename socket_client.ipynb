This is a simple client application using Python `socket` module in standart library. We start by importing that.

```python
import socket
```

To use the socket module first we are creating a socket object by calling:

```python
s = socket.socket()
```

## Creating A Client Application

Secondly, we need the address of the remote where the address is a tuple of two items:
- 1st is a string that is the ip address or domain name of the remote.
- 2nd is an integer between $0$ ond $2^{16}-1$ that is the port number on that server.

```python
remote = '172.16.2.5', 8080
remote = 'localhost', 9001
```

If you want to connect yourself (another application running on the same computer) then you can use `'127.0.0.1'` as IP address or `'localhost'` for domain name.

```python
s.connect(remote)
```

```python
s.sendall('Hello, world!'.encode())
```

```python
while True:
  try:
    print(s.revc().decode())
    s.sendall(input().encode())
  except KeyboardInterrupt:break
```

```python
s.shutdown(socket.SHUT_WR)
```
